```python
import pyterrier as pt
import pandas as pd
import os
import numpy as np

from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
```

```python
if not pt.started():
    pt.init()
```

```python
df_filtered = pd.read_csv('kitchen_filered_nan_new.csv')
df_filtered['docno'] = df_filtered['docno'].astype(str)
df_filtered = df_filtered[df_filtered['productName'].astype(str).map(len) > 10]
```

```python
# df_filtered.to_csv('kitchen_filered_nan_new.csv', index=False)
```

```python
pt_index_path = "./kitchen_index"

if not os.path.exists(pt_index_path + "/data.properties"):

    indexer = pt.DFIndexer(pt_index_path, overwrite=True)
    index_ref = indexer.index(df_filtered['reviewText'],df_filtered['docno'], df_filtered['summary'], df_filtered['productName'])
else:
    # if you already have the index, create an IndexRef from the data in pt_index_path
    # that we can use to load using the IndexFactory
    index_ref = pt.IndexRef.of(pt_index_path + "/data.properties")
    
index = pt.IndexFactory.of(index_ref)
```

```python
print(index.getCollectionStatistics().toString())
```

```python
query_df = pd.read_csv('query_qrels.csv')
query_df
```

```python
topics = query_df[['qid', 'query']]
qrels = query_df[['qid', 'docno', 'label']]
```

```python
topics = topics.drop_duplicates(ignore_index=True)
```

```python
qrels.fillna(1)
qrels['label'] = qrels['label'].astype('int64')
```

```python
arr = qrels['label']

arr
for i in range(len(arr)):
  c = qrels['label'][i]
  if np.isnan(c):
    print(c)
    print(i)
  elif not np.isfinite(c):
    print(c)
    print(i)
```

```python
topics
```

```python
qrels_copy = qrels.copy()
qrels_copy['docno'] = qrels_copy['docno'].astype(str)
qrels_copy
```

```python
SEED=10


tr_va_topics, test_topics = train_test_split(topics, test_size=4, random_state=SEED)
train_topics, valid_topics =  train_test_split(tr_va_topics, test_size=3, random_state=SEED)

```

```python
bm25 = pt.BatchRetrieve(index, wmodel="BM25") 
tf_idf = pt.BatchRetrieve(index, wmodel="TF_IDF") 
```

```python
def naive_weighting(keyFreq, posting, entryStats, collStats):

    
    # numDoc = collStats.getNumberOfDocuments()
    avgdocLen = collStats.getAverageDocumentLength()
    
    tmp = posting.getDocumentLength()/ avgdocLen

    return tmp

naive_model =  lambda keyFreq, posting, entryStats, collStats: naive_weighting(keyFreq, posting, entryStats, collStats)
naive = pt.BatchRetrieve(index, wmodel=naive_model)
```

```python
# learning to rank
RANK_CUTOFF = 30

ltr_feats = (bm25 % RANK_CUTOFF) >> pt.text.get_text(index, ["summary", "productName"]) >> (
    pt.transformer.IdentityTransformer()
    **
    pt.BatchRetrieve(index, wmodel="CoordinateMatch")
    ** 
    (pt.text.scorer(body_attr="summary", takes='docs', wmodel='BM25') ) 
    **
    (pt.text.scorer(body_attr="productName", takes='docs', wmodel='BM25') ) 
)
```

```python
# ltr_feats.search('cheese fruits quick')
```

```python
rf = RandomForestRegressor(n_estimators=500, verbose=1, random_state=SEED, n_jobs=8)

rf_pipe = ltr_feats >> pt.ltr.apply_learned_model(rf)

%time rf_pipe.fit(train_topics, qrels_copy)
```
